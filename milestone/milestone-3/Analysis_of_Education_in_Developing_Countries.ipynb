```python
# visualize the correlation between total completion rate and literacy rate from the dataframe 

# create a list of relevant data
relevant_columns = ['ISO3', 'Countries and areas', 'Region', 'Sub-region', 'Development Regions', 
                    'P_Total',  'LS_Total', 'US_Total', 
                    'Youth_Literacy_Rate_Total', 'Adult_Literacy_Rate_Total']

# subset the dataframe
df_subset = developing_countries[relevant_columns]

# calculate correlation matrix
correlation_matrix = df_subset.corr(numeric_only=True)

# create correlation heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of completion rate and literacy rate')
plt.show()

```

```python
# visualize the correlation between total out of school rate and literacy rate from the dataframe 

# create a list of relevant data
relevant_columns = ['ISO3', 'Countries and areas', 'Region', 'Sub-region', 'Development Regions', 
                    'OOR_Primary_Total', 'OOR_LowerSecondary_Total', 'OOR_UpperSecondary_Total', 
                    'Youth_Literacy_Rate_Total', 'Adult_Literacy_Rate_Total']

# subset the dataframe
df_subset = developing_countries[relevant_columns]

# calculate correlation matrix
correlation_matrix = df_subset.corr(numeric_only=True)

# create correlation heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of out of school rate and literacy rate')
plt.show()

```

```python
developing_countries['P_Total'].corr(developing_countries['Youth_Literacy_Rate_Total'])
```

## Findings 

As we can see here, the total completion rate in primary level, lower secondary level, and upper secondary level are positively correlated with the total youth and adult literacy rate. On the other hand, the out of school rate in all there levels are negatively correlated to the total youth and adult literacy rate. Suprisingly, the total youth literacy rate is strongly correlated with the total adult literacy rate.

Overall, we can conclude that the more students finish their education levels in a developing nation, the greater the literacy rate. The growth in out-of-school rates across all education levels will harm a developing country's literacy rate. Finally, a higher literacy rate among kids can increase the number of literate adults.

In conclusion, we can observe many females are loosing their right to access education in most developing countries.
