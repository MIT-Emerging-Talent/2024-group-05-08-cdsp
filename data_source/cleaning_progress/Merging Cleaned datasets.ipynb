```python
# First, merge df1 and df2
combined_OOR_CR = pd.merge(OOR, CR, 
                          on=['ISO3', 'Countries and areas', 'Region', 'Sub-region', 'Development Regions'], 
                          how='outer')

# Then, merge the result with LR
combined_df_OOR_CR_LR = pd.merge(combined_OOR_CR, LR, 
                           on=['ISO3', 'Countries and areas', 'Region', 'Sub-region', 'Development Regions'],  
                           how='outer')

# Finally, merge the result with AR
final_combined_df = pd.merge(combined_df_OOR_CR_LR, ATR, 
                             on=['ISO3', 'Countries and areas', 'Region', 'Sub-region', 'Development Regions'],
                             how='outer')
```
