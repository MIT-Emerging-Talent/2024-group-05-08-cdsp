```python
import numpy as np
import pandas as pd
```

```python
excel_file_path = r"C:\Users\tasab\Downloads\Education data\School-Age-Digital-Connectivity.xlsx"
```

```python
all_sheets = pd.read_excel(excel_file_path, sheet_name = None)
```

```python
Total_school_age = all_sheets['Total school age']
```

```python
df = Total_school_age.copy()
df.head()
```

```python
column_names = ['ISO3', 'Countries and areas', 'Region', 'Sub-region',
       'Income Group', 'Total', 'Rural', 'Urban', 'Poorest',
       'Richest', 'Data source', 'Time period']
```

```python
df.columns.values
```

```python
df['Countries and areas'].size
```

```python
df.columns = column_names
```

```python
df.drop(0,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
columns_to_convert = ['Rural', 'Urban', 'Poorest', 'Richest']
df[columns_to_convert] = df[columns_to_convert].astype(float)
df['Time period'] = pd.to_datetime(df['Time period'], errors='coerce')
df.info()
```

```python
df.shape
```

```python
df.size
```

```python
df_with_null = df[df.isnull().any(axis=1)]
```

```python
df_with_null
```

```python
df_with_null.size
```

```python
# Impute missing values with the mean of each column
#This is a simple method and can work well if the missing values are missing completely at random.
df['Rural'].fillna(df['Rural'].mean(), inplace=True)
df['Urban'].fillna(df['Urban'].mean(), inplace=True)
df['Poorest'].fillna(df['Poorest'].mean(), inplace=True)
df['Richest'].fillna(df['Richest'].mean(), inplace=True)
```
